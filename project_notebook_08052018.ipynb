### Preprocessing

```python
# Converting the date data to Date type
dt['datetime'] = pd.to_datetime(dt['datetime'], format="%A, %m/%d/%y %H:%M:%S %p")
dt['user_aliasborndate'] = pd.to_datetime(dt['user_aliasborndate'], format="%m/%d/%y")
dt['date'] = dt['datetime'].dt.date.astype('datetime64[ns]')
```

* Not many posts with empty text. I will just remove the posts because they don't seem very meaningful posts. 

```python
dt.text = dt.text.fillna('')
print(sum(dt['text'] == '')/dt.shape[0])
```

```python
dt.loc[(dt['text'] == '')]
```

```python
dt = dt.loc[dt['text'] != ""]
```

* Getting Sentiment for each of the texts:  

```python
import re

dt['text2'] = dt['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
dt['text2'] = dt['text2'].apply(lambda x: re.sub('http\S*', '', x))
dt['text2'] = dt['text2'].apply(lambda x: re.sub('\s+', ' ', x))
```

```python
dt.sample(5)
```

```python
dt.shape
```

* Filtering

```python
# Filtering - stopwords

from nltk.corpus import stopwords
stop = stopwords.words('english')
dt['text2'] = dt['text2'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
# Filtering - punctuation 
dt['text2'] = dt['text2'].apply(lambda x: re.sub('[^\w\s]','', x))
```

```python
# Tokenization
# from textblob import TextBlob
# dt['text2'] = dt['text2'].apply(lambda x: TextBlob()
```

```python
more_stopwords = ['apple', 'aapl']
#stop.extend(more_stopwords)
dt['text2'] = dt['text2'].apply(lambda x: " ".join(x for x in x.split() if x not in more_stopwords))
```

```python
def makeWordCloud(data):
    from wordcloud import WordCloud
    
    wc = WordCloud(background_color="white", 
               max_words=2000,
              width=1600,
              height=800)
    wc.generate(''.join(data['text2']))
    
    # show
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    #plt.figure(figsize=(20, 10))
    plt.axis("off")
    plt.savefig('wordcloud.png', dpi=300)
    plt.show()
```

### Pre-processing the data for the correlation analysis
* I assumed that **the number of the comments per day (increasing rate)** and **the sentiment per day** are the variables that could be used for predicting the stock price, so I pre-processed the data aggregating the number of the comments and the average sentiment score per day to get the correlation between the numbers and the historical stock price. 

```python
# Get the sentiment score for each of the texts
def sentiment_func(data):
    from textblob import TextBlob
    sentiment = TextBlob(data['text2'])
    data['polarity'] = sentiment.polarity
    data['subjectivity'] = sentiment.subjectivity
    return data
```

```python
dtt = dt.apply(sentiment_func, axis=1)
```

```python
dtt.head(3)
```

```python
dt2 = dtt.groupby('date').agg({
    'datetime': ['count'],
    'polarity': ['mean'],
    'subjectivity': ['mean']})
# dt.groupby('date').agg('count')['datetime'].to_frame().to_csv('a.csv')
dt2.columns = dt2.columns.droplevel(1)
dt2.head()
```

```python
dt2 = pd.merge(price_dt, dt2, how='left', left_index=True, right_index=True).rename(columns={'datetime':'numPosts'}).fillna(0)
```

```python
dt2['numPosts_rate'] = dt2['numPosts']/dt2['numPosts'].shift(1) - 1
dt2 = dt2.fillna(0)
dt2 = dt2.iloc[1:, :]
dt2.head()
```

```python
dt2.to_csv('dt5.csv')
```
